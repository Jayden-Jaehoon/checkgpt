```python
import os
from dotenv import load_dotenv
from langchain_teddynote import logging
from langchain_openai import ChatOpenAI

load_dotenv()
print(f"[API KEY]\n{os.environ['OPENAI_API_KEY']}")
# 프로젝트 이름을 입력합니다.
logging.langsmith("CheckGPT")
```

```python
# 객체 생성
llm1 = ChatOpenAI(
    temperature=0.1,  # 창의성 (0.0 ~ 2.0)
    model_name="gpt-4o",  # 모델명
)
# 질의내용
question = "주식시장을 모형화하기 위해서 100개의 agent를 만드려고 하는데, 각 agent마다 prmopt를 다르게 작성하고 싶어 우선 10개의 서로다른 prompt를 작성해줄 수 있니"
answer1 = llm1.stream(question)
# 스트리밍 방식으로 각 토큰을 출력합니다. (실시간 출력)
for token in answer1:
    print(token.content, end="", flush=True)
```

```python
from langchain.chat_models import ChatOpenAI

# LLM 객체 설정
llm1 = ChatOpenAI(
    temperature=0.1,  # 창의성 조정
    model_name="gpt-4o"  # 사용할 모델명
)

# 투자자 에이전트 딕셔너리
investment_agents = {
    "value_investor": {
        "description": "당신은 장기적인 가치 투자자입니다. 기업의 재무제표를 분석하고, 저평가된 주식을 찾아내어 장기 보유를 목표로 합니다.",
        "task": "현재 시장에서 저평가된 10개의 주식을 찾아내세요"
    },
    "momentum_investor": {
        "description": "당신은 모멘텀 투자자입니다. 최근 상승세를 보이는 주식을 식별하고, 그 추세가 지속될 것으로 예상되는 주식에 투자합니다.",
        "task": "최근 3개월 동안 가장 높은 상승률을 보인 10개의 주식을 찾아내세요"
    },
    "dividend_investor": {
        "description": "당신은 배당 수익을 중시하는 투자자입니다. 안정적인 배당을 지급하는 기업을 찾아내고, 배당 수익률이 높은 주식에 투자합니다.",
        "task": "현재 배당 수익률이 가장 높은 10개의 주식을 찾아내세요"
    },
    "technical_analyst": {
        "description": "당신은 기술적 분석가입니다. 차트와 기술적 지표를 활용하여 매수 및 매도 시점을 결정합니다.",
        "task": "현재 시장에서 기술적 지표가 매수 신호를 보내고 있는 10개의 주식을 찾아내세요"
    },
    "quant_trader": {
        "description": "당신은 퀀트 트레이더입니다. 수학적 모델과 알고리즘을 사용하여 시장의 비효율성을 찾아내고, 이를 기반으로 거래 전략을 수립합니다.",
        "task": "현재 시장에서 비효율성을 발견할 수 있는 데이터를 분석하여 10개의 주식을 찾아내세요."
    },
    "risk_averse_investor": {
        "description": "당신은 리스크 회피형 투자자입니다. 안정적이고 변동성이 낮은 자산에 투자하며, 포트폴리오의 리스크를 최소화하는 것을 목표로 합니다.",
        "task": "현재 시장에서 가장 안정적인 10개의 주식을 찾아내세요"
    },
    "growth_investor": {
        "description": "당신은 성장주 투자자입니다. 높은 성장 잠재력을 가진 기업에 투자하며, 주가 상승을 통해 수익을 추구합니다.",
        "task": "현재 시장에서 가장 높은 성장 잠재력을 가진 10개의 주식을 찾아내세요"
    },
    "short_term_trader": {
        "description": "당신은 단기 트레이더입니다. 하루 또는 며칠 내에 매수와 매도를 반복하며, 작은 가격 변동을 통해 수익을 추구합니다.",
        "task": "오늘 시장에서 단기적으로 수익을 낼 수 있는 10개의 주식을 찾아내세요"
    },
    "socially_responsible_investor": {
        "description": "당신은 사회적 책임을 중시하는 투자자입니다. 환경, 사회, 지배구조(ESG) 기준을 충족하는 기업에 투자합니다.",
        "task": "현재 시장에서 ESG 점수가 높은 10개의 주식을 찾아내세요"
    }
}

# 에이전트별 프롬프트를 생성하고 LLM에 요청하는 함수
def get_responses(agents_dict, question):
    responses = {}
    for agent_name, agent_data in agents_dict.items():
        prompt = f"{agent_data['description']}\n\n{agent_data['task']}"
        print(f"*** {agent_name.upper()} 에이전트의 답변 ***")
        answer = llm1.stream(prompt)
        for token in answer:
            print(token.content,  end="", flush=True)  # 스트리밍 방식으로 실시간 출력
        responses[agent_name] = answer  # 각 에이전트의 답변 저장
        print("\n")
    return responses

# 질문을 바탕으로 모든 에이전트의 답변을 얻음
responses = get_responses(investment_agents, question="한국에서 미래전망이 좋은 주식 10종목을 추천해줘")
```
